## 충전소별 시간대별로 사용횟수 분석
- 분 단위로 분석 
- 하루 24시간을 1440분으로 구분 

### 1. 읽어들일 파일명 지정 ==> 수정필요

```python
# 수정필요
########### file_name 선언 ##################################
충전이력셋 = '남양주_충전이력_전처리.csv'
#############################################################
```

### 2. 이하 맨끝까지 수정 불필요

```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
%matplotlib inline
%matplotlib inline

#한글깨지는문제 보완
from matplotlib import font_manager,rc
font_location="c:/windows/fonts/malgun.ttf"
font_name=font_manager.FontProperties(fname=font_location).get_name()
rc('font',family=font_name)


################## 충전이력 CSV 읽기 ####################################################
충전이력 = pd.read_csv(f'{충전이력셋}', header=0, encoding='cp949', engine='python')
충전이력 = 충전이력.loc[:,['충전소명', 'start', 'end']]
충전이력 = 충전이력.sort_values(by='충전소명')

# Null 이 있을 수 있는 칼럼은 Null을 0로 변경
충전이력.fillna(0, inplace=True)

# 문자를 date 타입으로 변환
충전이력['start'] = pd.to_datetime(충전이력['start'], format='%Y-%m-%d %H:%M:%S')
충전이력['end'] = pd.to_datetime(충전이력['end'], format='%Y-%m-%d %H:%M:%S')

display(" 읽은 파일 한건만 조회", 충전이력.head(1))
################## 충전이력 CSV 읽기 끝###################################################
```

```python
충전소리스트 = 충전이력.충전소명.unique().tolist()
충전소갯수 = len(충전소리스트)
print (f"충전소는 총 {충전소갯수}개 이며, 충전소명은\n\n", 충전소리스트)

# #시작일시/종료일시의 시간과 분을 분단위로 변환 (예:02:10 ==> 130)
start = [] # 빈 리스트 생성
end = []
df_view = 충전이력.copy()

for i in range(len(df_view)):  
    dtstart = df_view.iloc[i,1]        # df의 1번째(실제는 두번째임, 파이썬은 0번째가 첫번째) 칼럼인 시작일시 칼럼
    dtend = df_view.iloc[i,2]          # df의 2번째(실제는 세번째임, 파이썬은 0번째가 첫번째) 칼럼인 종료일시 칼럼
    start1 = int(dtstart.strftime("%H")) * 60 + int(dtstart.strftime("%M"))  # 시간 * 60분
    end1 = int(dtend.strftime("%H")) * 60 + int(dtend.strftime("%M"))        # 더히시 분
    
    # 리스트에 누적
    start.append(start1)
    end.append(end1)
    
# start와 end 리스트 값들을 원본 데이터프레임에 합치기
df_view['start2'] = start
df_view['end2'] = end
display (df_view.head(1))
```

### 하루 00:00 ~ 23:59 을 1440 개로 구분 (24시간 * 60분 = 1440분)
- 1440 칼럼짜리 데이터프레임을 만든후
- 분당 이용 빈도 수를 누적

```python
# 충전소 갯수만큼, 1440분 짜리 빈 데이블 생성
cols = list(range(0, 1440, 1))  #0부터 1440(비포함, 즉 1339까지) 1씩 증가
ind = 충전소리스트
df_time = pd.DataFrame(0,index=ind,columns=cols) 

# start보다 end 가 작은 경우 별도 처리가 필요함 (즉 23:50 분 시작, 익일 00시 10분 종료 하는 경우)
df_view['정상'] = np.where((df_view['start2'] > df_view['end2']),False,True )  # start > end 조건이면 False, 아니면 True
df_정상 = df_view[df_view['정상']==True]   # True만 뽑아 정상으로 저장
df_예외 = df_view[df_view['정상']==False]  # Flase 만 뽑아 예외로 저장


#1440분 짜리 빈 테이블에 이용 빈도 누적 계산
for j, name in enumerate(충전소리스트):
    
    df_ = df_정상[df_정상['충전소명'] == name]
    for i in range(len(df_)): 
        s = df_.iloc[i,3]
        e = df_.iloc[i,4]
        if (s == e):
            df_time.iloc[j, s] += 1
        else:       
            df_time.iloc[j, s:e+1] += 1
        
        
    df_ = df_예외[df_예외['충전소명'] == name]
    for i in range(len(df_)): 
        s = df_.iloc[i,3]
        e = df_.iloc[i,4]
        df_time.iloc[j, s:1440] += 1
        df_time.iloc[j, 0:e+1] += 1

# 옆으로 길게 되어 있는 모얄을 밑으로 길게 전치
df_t = df_time.T
display("충전소별 사용빈도", df_t)
```
